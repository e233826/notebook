```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
from sklearn import preprocessing
adult_census_income = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', 
                                  names = ["Age", "Workclass","fnlwgt","education", "education-num","marital-status",
                                           "occupation","relationship","race", "sex", "capital.gain","capital.los",
                                           "hours.per.week","native.country", "income"] , na_values=["?"])
adult_census_income.info()
```

```
#checking missing valus since in dataset I saw that the unknown values shown via ? I replace with null
adult_census_income.replace('[?]', np.nan, regex=True , inplace=True)
adult_census_income
```

```

adult_census_income['Workclass'].replace( np.nan, 'Private' ,regex=True , inplace=True)
```

```
adult_census_income['occupation'].describe()
```

```

adult_census_income['occupation'].replace( np.nan, 'Prof-specialty' ,regex=True , inplace=True)
```

```
adult_census_income['native.country'].replace( np.nan, 'United-States' ,regex=True , inplace=True)
```

```

adult_census_income.isnull().sum()
```

```
###############################################preparing dataset for ML ######################################################
adult_census_income['income'].replace('<=50K','0',regex=True, inplace = True)
adult_census_income['income'].replace('>50K','1',regex=True,inplace = True)
#in Workclass colums to reduce variables
self_employed = ['Self-emp-not-inc','Self-emp-inc']
govt_employees = ['Local-gov','State-gov','Federal-gov']
adult_census_income['Workclass'].replace(to_replace = self_employed , value = 'Self_employed',regex=True,inplace = True)
adult_census_income['Workclass'].replace(to_replace = govt_employees, value = 'Govt_employees',regex=True,inplace = True)
#in education colum to reduce variables 
hs_grad = ['HS-grad','11th','10th','9th','12th']
elementary = ['1st-4th','5th-6th','7th-8th']
adult_census_income['education'].replace(to_replace = hs_grad,value = 'HS-grad',regex=True, inplace = True)
adult_census_income['education'].replace(to_replace = elementary,value = 'elementary_school',regex=True,inplace = True)
#in marital.status colum to reduce variable 
married= ['Married-spouse-absent','Married-civ-spouse','Married-AF-spouse']
separated = ['Separated','Divorced']
adult_census_income['marital-status'].replace(to_replace = married , value = 'Married',regex=True, inplace = True)
adult_census_income['marital-status'].replace(to_replace = separated,value = 'Separated',regex=True, inplace = True)
#since some of the colums is kind of dublicate or extra data I can be deleted 
#Education num and education are giving similar information also Relationship and marital status imply similar information.Hence keeping only one of the two
adult_census_income.drop('capital.gain' ,axis = 1,inplace = True)
adult_census_income.drop('capital.los' ,axis = 1,inplace = True)
adult_census_income.drop('education-num' , axis = 1, inplace = True)
adult_census_income.drop('relationship' , axis = 1, inplace = True)
adult_census_income.drop('fnlwgt' , axis = 1, inplace = True)
from sklearn.preprocessing import LabelEncoder
lb_make = LabelEncoder()
adult_census_income['Workclass'] = lb_make.fit_transform(adult_census_income['Workclass'])
adult_census_income['education'] = lb_make.fit_transform(adult_census_income['education'])
adult_census_income['marital-status'] = lb_make.fit_transform(adult_census_income['marital-status'])
adult_census_income['race'] = lb_make.fit_transform(adult_census_income['race'])
adult_census_income['sex'] = lb_make.fit_transform(adult_census_income['sex'])
adult_census_income['occupation'] = lb_make.fit_transform(adult_census_income['occupation'])
adult_census_income['native.country'] = lb_make.fit_transform(adult_census_income['native.country'])
adult_census_income
```

```
data = adult_census_income
target = data['income']
del data['income']
data = pd.get_dummies(data)

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data, target, train_size = 0.7)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,BaggingClassifier,ExtraTreesClassifier
#from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_curve, auc
import warnings
import time
import datetime
warnings.filterwarnings('ignore')
```

```
clf_knn = KNeighborsClassifier(n_neighbors=6,metric='minkowski',p=2)
clf_log = LogisticRegression()
clf_tree = DecisionTreeClassifier()
clf_SVC = SVC ()
clf_forest = RandomForestClassifier()
models = {clf_log:'Logistic Regressions',clf_tree:'Decision Tree', clf_forest:' Random Forest ',clf_log:'KNN' }
#           , clf_knn:'KNN',clf_tree:'Decision Tree',clf_SVC:'SVC' }

def training (algo,name, X_train, X_test, Y_train, Y_test):
    algo.fit(X_train,Y_train)
    y_pred = algo.predict(X_test)
    print (f"***************************{name}***************************")
    print (f"acuracy score is {accuracy_score(Y_test,y_pred)}")
    return Y_test, y_pred

def result (Y_test, y_pred):
    print ("Confusion Matrix")
    cm=confusion_matrix(Y_test, y_pred)
    print (cm) 
    TP=cm[1,1]
    TN=cm[0,0]
    FP=cm[0,1]
    FN=cm[1,0]
    from sklearn import metrics
    accuracy = metrics.accuracy_score(Y_test, y_pred)
    f1_score =  100*((2*TP) / (2*TP + FP + FN)).round(4)
    precision = 100*((TP) / (TP+FP)).round(4)
    recall = 100*((TP) / (TP+FN)).round(4)
    print (f"Accuracy is: {accuracy}%\nF1 score is: {f1_score}%\nPrecision is: {precision}%\nRecall is: {recall}%")

def main (models):
    acc_score = []
    for algo,name in models.items(): 
        start_time = time.time()
        Y_test_train,y_pred = training (algo,name,X_train, X_test,Y_train, Y_test)
        result(Y_test_train,y_pred)
        acc_score.append(result)
        print (f"Elapsed time: {datetime.timedelta(seconds=time.time()-start_time)} ")
    return acc_score
        

main (models)
    
```

```
############################################Decision tree#####################################################3
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state=1000, max_depth=5)
dtc = dtc.fit(X_train, Y_train)
predicted_dtc = dtc.predict(X_test)
from sklearn import metrics
metrics.accuracy_score(Y_test, predicted_dtc)
```

```
####################################Cross validation Decision tree ####################################################
from sklearn.model_selection import cross_val_score  
all_accuracies = cross_val_score(estimator=dtc, X=X_train, y=Y_train, cv=5)
all_accuracies.mean()
```

```
############################################Grid search#######################################################
from sklearn.model_selection import GridSearchCV
param_grid = {"criterion": ["gini", "entropy"], "min_samples_split": [2, 10], "max_depth": [2, 5, 10]}
grid = GridSearchCV(dtc, param_grid, cv=3, scoring='neg_mean_squared_error')
grid.fit(X_train, Y_train.values.ravel())
y_pred_class = grid.predict(X_test)
grid.best_params_
```
