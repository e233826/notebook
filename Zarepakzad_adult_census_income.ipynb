```
############################## IMPORTING DATA SET ##################################
####################################################################################

import pandas as pd
import numpy as np
from sklearn import preprocessing
income_census = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',names = ["age", "workclass","fnlwgt","education", "education-num","marital-status",                                           "occupation","relationship","race", "sex", "capital.gain","capital.los" ,
"hours.per.week","native.country", "income"] , na_values=["?"])
income_census
```

```
income_census.info()
```

```
############################ CHECKING EMPTY INPUTS #################################
####################################################################################

income_census.replace('[?]', np.nan, regex=True , inplace=True)
income_census
```

```
############################ COUNTING EMPTY INPUTS #################################
####################################################################################

income_census.isnull().sum()
```

```
####### REPLACING EMPTY INPUTS OF "workclass" ATTRIBUTE WITH ITS MODE ##############
####################################################################################

income_census['workclass'].mode()
```

```
income_census ['workclass'].replace( np.nan, 'Private',regex=True , inplace=True)
```

```
##### REPLACING EMPTY INPUTS OF "occupation" ATTRIBUTE WITH ITS MODE ###############
####################################################################################

income_census['occupation'].mode()
```

```
income_census ['occupation'].replace( np.nan, ' Prof-specialty',regex=True , inplace=True)
```

```
##### REPLACING EMPTY INPUTS OF "native.country" ATTRIBUTE WITH ITS MODE ###########
####################################################################################

income_census['native.country'].mode()
```

```
income_census['native.country'].replace( np.nan, 'United-States',regex=True , inplace=True)
```

```
################### COUNTING EMPTY INPUTS AFTER REPLACEMENT ########################
####################################################################################

income_census.isnull().sum()
```

```
from sklearn.preprocessing import LabelEncoder
lb_make = LabelEncoder()
```

```
######## MERGING SIMILAR CHARACTERISTICS OF "workclass" & LABEL ENCODING ###########
####################################################################################

self_emp= ['Self-emp-not-inc','Self-emp-inc']
gov_emp= ['Local-gov','State-gov','Federal-gov']
no_earn= ['Without-pay',' Never-worked']
income_census['workclass'].replace(to_replace = self_emp , value = 'self_emp',regex=True , inplace=True)
income_census['workclass'].replace(to_replace = gov_emp, value = 'gov_emp',regex=True , inplace=True)
income_census['workclass'].replace(to_replace = no_earn, value = 'no_earn',regex=True , inplace=True)
income_census['workclass']= lb_make.fit_transform(income_census['workclass'])
income_census.head(5)
```

```
######## MERGING SIMILAR CHARACTERISTICS OF "education"  & LABEL ENCODING ##########
####################################################################################

h_school = ['HS-grad','11th','10th','9th','12th']
e_school = ['1st-4th','5th-6th','7th-8th']
university = ['Bachelors','Some-college']
income_census['education'].replace(to_replace = h_school,value = 'h_school',regex=True, inplace = True)
income_census['education'].replace(to_replace = e_school,value = 'e_school',regex=True,inplace = True)
income_census['education'].replace(to_replace = university,value = 'university',regex=True,inplace = True)
income_census['education']= lb_make.fit_transform(income_census['education'])
income_census.head(5)
```

```
###### MERGING SIMILAR CHARACTERISTICS OF "marital-status"  & LABEL ENCODING #######
####################################################################################

married= ['Married-spouse-absent','Married-civ-spouse','Married-AF-spouse']
separated = ['Separated','Divorced']
income_census['marital-status'].replace(to_replace = married , value = 'married',regex=True, inplace = True)
income_census['marital-status'].replace(to_replace = separated,value = 'separated',regex=True, inplace = True)
income_census['marital-status']= lb_make.fit_transform(income_census['marital-status'])
income_census.head(5)
```

```
####################### LABEL ENCODING OBJECT ATTRIBUTES ###########################
####################################################################################

income_census['race'] = lb_make.fit_transform(income_census['race'])
income_census['sex'] = lb_make.fit_transform(income_census['sex'])
income_census['occupation'] = lb_make.fit_transform(income_census['occupation'])
income_census['native.country'] = lb_make.fit_transform(income_census['native.country'])
income_census['income'].replace('<=50K','0',regex=True, inplace = True)
income_census['income'].replace('>50K','1',regex=True,inplace = True)
```

```
################### DROPPING DUPLICATED/SIMILAR ATTRIBUTES #########################
####################################################################################

income_census.drop('fnlwgt' , axis = 1, inplace = True)
income_census.drop('education-num' , axis = 1, inplace = True)
income_census.drop('relationship' , axis = 1, inplace = True)
income_census.drop('capital.gain' ,axis = 1,inplace = True)
income_census.drop('capital.los' ,axis = 1,inplace = True)
```

```
###################### PREPARING TRAIN AND TEST SETS ###############################
####################################################################################

from sklearn.model_selection import train_test_split
data = income_census
target = data['income']
del data['income']
data = pd.get_dummies(data)
X_train, X_test, Y_train, Y_test = train_test_split(data, target, test_size = 0.2)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```
########### IMPORTING LEARNING/PEFORMANCE EVALUATION LIBRARIES #####################
####################################################################################

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score,GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn import metrics
```

```
######################### a.1. LOGISTIC REGRESSION:ACCURACY ########################
####################################################################################

LR = LogisticRegression(solver = 'sag', max_iter=10000, C =0.01)
LR = LR.fit(X_train,Y_train)
predicted_LR = LR.predict(X_test)
metrics.accuracy_score(Y_test, predicted_LR)
```

```
##################### a.2. LOGISTIC REGRESSION:CROSS-VALIDATION ####################
####################################################################################

all_accuracies = cross_val_score(estimator=LR, X=X_train, y=Y_train, cv=3)
all_accuracies.mean()
```

```
####################### a.3. LOGISTIC REGRESSION:GRID SEARCH #######################
####################################################################################

param_grid = {'C' :np.arange(0.01,1, 0.2), 'solver': ["lbfgs", "sag"]}
grid = GridSearchCV(LR, param_grid, cv = 3)
grid.fit(X_train, Y_train)
grid.best_params_
```

```
############################ b.1. DECISION TREE:ACCURACY ###########################
####################################################################################

DT = DecisionTreeClassifier(max_depth=10, min_samples_split= 12)
DT = DT.fit(X_train, Y_train)
predicted_DT = DT.predict(X_test)
metrics.accuracy_score(Y_test, predicted_DT)
```

```
######################### b.2. DECISION TREE:CROSS-VALIDATION ######################
####################################################################################

all_accuracies = cross_val_score(estimator=DT, X=X_train, y=Y_train, cv=3)
all_accuracies.mean()
```

```
########################## b.3. DECISION TREE:GRID SEARCH ##########################
####################################################################################

param_grid = {'min_samples_split':np.arange(4,14,2), 'max_depth':np.arange(4,14,2)}
grid = GridSearchCV(DT, param_grid, cv=3, scoring='neg_mean_squared_error')
grid.fit(X_train, Y_train.values.ravel())
y_pred_class = grid.predict(X_test)
grid.best_params_
```

```
##################### c.1. SUPPPORT VECTOR MECHINE:ACCURACY ########################
####################################################################################

SV = SVC(gamma='auto', C=0.8)
SV =SV.fit(X_train, Y_train)
predicted_SV = SV.predict(X_test)
metrics.accuracy_score(Y_test, predicted_SV)
```

```
############### c.2. SUPPPORT VECTOR MECHINE:CROSS-VALIDATION  #####################
####################################################################################

all_accuracies = cross_val_score(estimator=SV, X=X_train, y=Y_train, cv=3)
all_accuracies.mean()
```

```
######################## c.3. SUPPPORT VECTOR: GRID SEARCH #########################
####################################################################################

import warnings
warnings.filterwarnings('ignore')
param_grid = {'C' :[0.2, 0.5, 0.8]}
SV = SVC()
grid = GridSearchCV(SV, param_grid, cv=3)
grid.fit(X_train, Y_train)
grid.best_params_
```

```
######################## d.1. K-NEAREST NEIGHBOR: ACCURACY  ########################
####################################################################################

KNN = KNeighborsClassifier(n_neighbors= 9, weights= 'uniform')
KNN =KNN.fit(X_train, Y_train)
predicted_KNN = KNN.predict(X_test)
metrics.accuracy_score(Y_test, predicted_KNN)
```

```
#################### d.2. K-NEAREST NEIGHBOR: CROSS-VALIDATION  ####################
####################################################################################

all_accuracies = cross_val_score(estimator=KNN, X=X_train, y=Y_train, cv=3)
all_accuracies.mean()
```

```
###################### d.3. K-NEAREST NEIGHBOR: GRID SEARCH ########################
####################################################################################

param_grid = {'n_neighbors' :np.arange(1,10,2), 'weights':["uniform","distance"]}
grid = GridSearchCV(KNN, param_grid, cv=3)
grid.fit(X_train, Y_train)
grid.best_params_
```
